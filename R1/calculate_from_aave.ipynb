```python
# -----------------------------------------------------------------------------
# 步骤 1: 导入必要的库并定义常量
# -----------------------------------------------------------------------------
import requests
import json
from decimal import Decimal

# 严格按照 PDF 定义的稳定币列表
STABLECOIN_SYMBOLS = ["USDC", "USDT", "DAI", "FRAX", "LUSD", "GHO"]

# --- 2. 配置 CoinGecko API ---

# 使用你已验证的 API 密钥
API_KEY = 'CG-xtrHV9nWxQxtwq8tyThUtjNC' 
HEADERS = {
    'accept': 'application/json',
    'x-cg-demo-api-key': API_KEY
}
CG_BASE_URL = "https://api.coingecko.com/api/v3"

# (已修正) COIN_ID_MAP 现在只需要包含“基础”资产
# 我们将用 'ethereum' 和 'bitcoin' 的价格来代理其他衍生品
COIN_ID_MAP = {
    # 基础代理
    "ETH": "ethereum",
    "BTC": "bitcoin",
    
    # 稳定币 (根据您的 reserves 列表)
    "AAVE": "aave",
    "EURC": "euro-coin",
    "GHO": "gho",
    "USDC": "usd-coin",
    "USDbC": "bridged-usd-coin-base",
    
    # 其他非 ETH/BTC 资产 (如果需要的话)
    # (您的 reserves 列表中没有其他此类资产)
}

# --- 3. 定义代理资产组 ---
# 我们将使用 'ETH' 的价格来计算这些资产
ETH_PROXY_SYMBOLS = ["WETH", "weETH", "ezETH", "cbETH", "wstETH", "wrsETH"]
# 我们将使用 'BTC' 的价格来计算这些资产
BTC_PROXY_SYMBOLS = ["tBTC", "cbBTC", "LBTC"]


# -----------------------------------------------------------------------------
# 步骤 4: 您的 GraphQL 响应数据 (用于获取资产数量)
# -----------------------------------------------------------------------------
# 这是您在上一个请求中提供的 JSON 响应
response_data = {
  "data": {
    "reserves": [
      {
        "decimals": 18, "price": {"id": "0x04c0599ae5a44757c0af6f9ec3b93da8976c150a", "priceInEth": "0"},
        "reserveLiquidationThreshold": "7700", "symbol": "weETH", "totalCurrentVariableDebt": "5711413226611483813",
        "totalLiquidity": "101060475476073517363523", "totalPrincipalStableDebt": "0"
      },
      {
        "decimals": 18, "price": {"id": "0x236aa50979d5f3de3bd1eeb40e81137f22ab794b", "priceInEth": "9691461317600"},
        "reserveLiquidationThreshold": "7800", "symbol": "tBTC", "totalCurrentVariableDebt": "2319486623176828625",
        "totalLiquidity": "10340642374716541723", "totalPrincipalStableDebt": "0"
      },
      {
        "decimals": 18, "price": {"id": "0x2416092f143378750bb29b79ed961ab195cceea5", "priceInEth": "0"},
        "reserveLiquidationThreshold": "10", "symbol": "ezETH", "totalCurrentVariableDebt": "0",
        "totalLiquidity": "62724757656228962796", "totalPrincipalStableDebt": "0"
      },
      {
        "decimals": 18, "price": {"id": "0x2ae3f1ec7f1f5012cfeab0185bfc7aa3cf0dec22", "priceInEth": "0"},
        "reserveLiquidationThreshold": "7900", "symbol": "cbETH", "totalCurrentVariableDebt": "73211718930522736606",
        "totalLiquidity": "7275725911966508937708", "totalPrincipalStableDebt": "0"
      },
      {
        "decimals": 18, "price": {"id": "0x4200000000000000000000000000000000000006", "priceInEth": "322840000000"},
        "reserveLiquidationThreshold": "8300", "symbol": "WETH", "totalCurrentVariableDebt": "111935033899531092134493",
        "totalLiquidity": "135773819228238313273787", "totalPrincipalStableDebt": "0"
      },
      {
        "decimals": 6, "price": {"id": "0x60a3e35cc302bfa44cb288bc5a4f316fdb1adb42", "priceInEth": "0"},
        "reserveLiquidationThreshold": "7800", "symbol": "EURC", "totalCurrentVariableDebt": "14738394184678",
        "totalLiquidity": "28137923993177", "totalPrincipalStableDebt": "0"
      },
      {
        "decimals": 18, "price": {"id": "0x63706e401c06ac8513145b7687a14804d17f814b", "priceInEth": "18186543000"},
        "reserveLiquidationThreshold": "6500", "symbol": "AAVE", "totalCurrentVariableDebt": "0",
        "totalLiquidity": "5743469788964741943928", "totalPrincipalStableDebt": "0"
      },
      {
        "decimals": 18, "price": {"id": "0x6bb7a212910682dcfdbd5bcbb3e28fb4e8da10ee", "priceInEth": "0"},
        "reserveLiquidationThreshold": "0", "symbol": "GHO", "totalCurrentVariableDebt": "7831765623389494312280982",
        "totalLiquidity": "14562490328570051552693819", "totalPrincipalStableDebt": "0"
      },
      {
        "decimals": 6, "price": {"id": "0x833589fcd6edb6e08f4c7c32d4f71b54bda02913", "priceInEth": "99986468"},
        "reserveLiquidationThreshold": "7800", "symbol": "USDC", "totalCurrentVariableDebt": "270641188114918",
        "totalLiquidity": "394263161444332", "totalPrincipalStableDebt": "0"
      },
      {
        "decimals": 18, "price": {"id": "0xc1cba3fcea344f92d9239c08c0568f6f2f0ee452", "priceInEth": "0"},
        "reserveLiquidationThreshold": "7900", "symbol": "wstETH", "totalCurrentVariableDebt": "6871703749728494496329",
        "totalLiquidity": "27527287398506498472580", "totalPrincipalStableDebt": "0"
      },
      {
        "decimals": 8, "price": {"id": "0xcbb7c0000ab88b473b1f5afd9ef808440eed33bf", "priceInEth": "8155695500000"},
        "reserveLiquidationThreshold": "7800", "symbol": "cbBTC", "totalCurrentVariableDebt": "40231161143",
        "totalLiquidity": "253922858089", "totalPrincipalStableDebt": "0"
      },
      {
        "decimals": 6, "price": {"id": "0xd9aaec86b65d86f6a7b5b1b0c42ffa531710b6ca", "priceInEth": "100010000"},
        "reserveLiquidationThreshold": "7800", "symbol": "USDbC", "totalCurrentVariableDebt": "403423125522",
        "totalLiquidity": "437671384568", "totalPrincipalStableDebt": "0"
      },
      {
        "decimals": 8, "price": {"id": "0xecac9c5f704e954931349da37f60e39f515c11c1", "priceInEth": "10849218252036"},
        "reserveLiquidationThreshold": "7300", "symbol": "LBTC", "totalCurrentVariableDebt": "0",
        "totalLiquidity": "5859006953", "totalPrincipalStableDebt": "0"
      },
      {
        "decimals": 18, "price": {"id": "0xedfa23602d0ec14714057867a78d01e94176bea0", "priceInEth": "0"},
        "reserveLiquidationThreshold": "10", "symbol": "wrsETH", "totalCurrentVariableDebt": "0",
        "totalLiquidity": "6478564545149594243854", "totalPrincipalStableDebt": "0"
      }
    ]
  }
}

# -----------------------------------------------------------------------------
# 步骤 5: 从 CoinGecko 获取当前 USD 价格的函数
# -----------------------------------------------------------------------------

def get_current_prices_usd(coin_id_map: dict) -> dict:
    """
    从 CoinGecko 获取 COIN_ID_MAP 中所有资产的当前 USD 价格。
    返回一个 {Symbol: Price_USD} 格式的字典。
    """
    print("--- 正在从 CoinGecko 获取当前价格快照... ---")
    
    ids_string = ",".join(coin_id_map.values())
    url = f"{CG_BASE_URL}/simple/price"
    params = { 'ids': ids_string, 'vs_currencies': 'usd' }
    
    try:
        response = requests.get(url, headers=HEADERS, params=params)
        response.raise_for_status() 
        cg_data = response.json()
        
        price_map_usd = {}
        for symbol, coin_id in coin_id_map.items():
            if coin_id in cg_data and 'usd' in cg_data[coin_id]:
                price = cg_data[coin_id]['usd']
                price_map_usd[symbol] = Decimal(str(price)) 
                print(f"  -> 已获取 {symbol:<6} 价格: ${price_map_usd[symbol]:.4f} USD")
            else:
                print(f"  -> 警告: 未能在 CoinGecko 响应中找到 {symbol} ({coin_id}) 的价格。")
                
        print("--- 价格获取成功 ---")
        return price_map_usd
        
    except Exception as e:
        print(f"  -> 严重错误: CoinGecko API 请求失败或处理时出错: {e}")
        return None

# -----------------------------------------------------------------------------
# 步骤 6: 修正后的 R1 计算函数 (实施您的代理策略)
# -----------------------------------------------------------------------------

def calculate_r1_proxy_strategy(reserves_data: list, prices_usd_map: dict):
    """
    根据 Aave Subgraph 数量 和 CoinGecko USD 价格计算 R_I 指标。
    使用 ETH 和 BTC 价格作为其衍生品的代理。
    """
    
    numerator_R1 = Decimal(0) 
    denominator_R1 = Decimal(0) 

    # 从价格图中获取 ETH 和 BTC 的基础价格
    ETH_PRICE_USD = prices_usd_map.get("ETH")
    BTC_PRICE_USD = prices_usd_map.get("BTC")

    if not ETH_PRICE_USD or not BTC_PRICE_USD:
        print("!!! 严重错误: 未能获取 ETH 或 BTC 的基础价格。无法继续。")
        return None

    print("\n--- R1 计算过程 (基于 CoinGecko 代理策略) ---")

    for asset in reserves_data:
        symbol = asset.get('symbol')
        price_in_usd = None
        
        # --- 您的代理逻辑在此处实现 ---
        if symbol in STABLECOIN_SYMBOLS:
            price_in_usd = prices_usd_map.get(symbol)
        elif symbol in ETH_PROXY_SYMBOLS:
            price_in_usd = ETH_PRICE_USD
        elif symbol in BTC_PROXY_SYMBOLS:
            price_in_usd = BTC_PRICE_USD
        else:
            # 适用于 AAVE, EURC, USDbC 等其他资产
            price_in_usd = prices_usd_map.get(symbol)
        # ---------------------------------
            
        if not price_in_usd:
            print(f"  [跳过] {symbol}: 无法在 COIN_ID_MAP 中找到其自身或代理价格。")
            continue
            
        if asset.get('decimals') is None:
            print(f"  [跳过] {symbol}: 缺少 decimals 字段。")
            continue
            
        decimals = int(asset['decimals'])
        divisor = Decimal(10) ** decimals
        
        try:
            D_j_amount = Decimal(asset['totalLiquidity']) / divisor
            B_j_amount = (Decimal(asset['totalPrincipalStableDebt']) +
                          Decimal(asset['totalCurrentVariableDebt'])) / divisor
            tau_j = Decimal(asset['reserveLiquidationThreshold']) / Decimal(10000)
        except Exception as e:
            print(f"  [跳过] {symbol}: 在 D_j 或 B_j 转换时出错: {e}")
            continue
            
        # 核心分类逻辑 (使用 USD 价值)
        if symbol in STABLECOIN_SYMBOLS:
            # 稳定币 -> 贡献分子
            B_j_value_usd = B_j_amount * price_in_usd
            print(f"  [分子+] {symbol:<6}: 增加借款价值 ${B_j_value_usd:,.2f} USD")
            numerator_R1 += B_j_value_usd
        else:
            # 非稳定币 -> 贡献分母
            if tau_j == 0: 
                print(f"  [跳过] {symbol:<6}: 清算阈值为 0，不计入分母。")
                continue
                
            net_deposits_amount = D_j_amount - B_j_amount
            weighted_value_usd = (net_deposits_amount * tau_j) * price_in_usd
            
            # (打印时显示它使用的是哪个代理价格)
            proxy_note = ""
            if symbol in ETH_PROXY_SYMBOLS: proxy_note = " (使用 ETH 价格)"
            if symbol in BTC_PROXY_SYMBOLS: proxy_note = " (使用 BTC 价格)"
                
            print(f"  [分母+] {symbol:<6}: 增加风险加权净存款价值 ${weighted_value_usd:,.2f} USD{proxy_note}")
            denominator_R1 += weighted_value_usd

    # 最终计算
    print("-------------------")
    print(f"计算完成: ")
    print(f"  - 总分子 (稳定币总借款价值): ${numerator_R1:,.2f} USD")
    print(f"  - 总分母 (风险加权净存款价值): ${denominator_R1:,.2f} USD")
    
    if denominator_R1 <= 0:
        print("警告: 分母为 0 或负数，R1 视为无穷大 (极度危险)")
        return float('inf')
    
    R_I = numerator_R1 / denominator_R1
    return float(R_I)

# -----------------------------------------------------------------------------
# 步骤 7: 运行最终计算
# -----------------------------------------------------------------------------
if __name__ == "__main__":
    
    reserves_list = response_data.get("data", {}).get("reserves", [])
    
    if not reserves_list:
        print("错误: 未能在 'response_data' 中找到 'reserves' 数据。")
    else:
        # 1. 从 CoinGecko 获取基础价格
        prices_usd_map = get_current_prices_usd(COIN_ID_MAP)
        
        if prices_usd_map:
            # 2. 运行 R1 计算 (使用您的代理策略)
            r1_value = calculate_r1_proxy_strategy(reserves_list, prices_usd_map)
            
            if r1_value is not None:
                print("\n===============================")
                print(f" 最终 R1 (协议风险指标): {r1_value:.6f}")
                print("===============================")
                
                if 0.4 < r1_value < 1.0:
                     print("\n提示：结果在 (0.4 ~ 1.0) 的合理区间内。这看起来是一个健康、有效的值。")
                else:
                     print("\n提示：结果仍超出预期范围。请检查：")
                     print("1. STABLECOIN_SYMBOLS 列表是否符合您的策略定义？")
                     print("2. ETH/BTC 代理组是否覆盖了所有相关资产？")
        else:
            print("由于无法获取 CoinGecko 价格，已停止 R1 计算。")
```

```python
import pandas as pd
import requests
import json
from decimal import Decimal
from pathlib import Path
from tqdm import tqdm
import os # 确保 os 已导入

# -----------------------------------------------------------------------------
# 步骤 1: 重复的定义（确保此单元格可独立运行所需配置）
# -----------------------------------------------------------------------------

# 严格按照 PDF 定义的稳定币列表
STABLECOIN_SYMBOLS = ["USDC", "USDT", "DAI", "FRAX", "LUSD", "GHO"]

# (已修正) COIN_ID_MAP 现在只需要包含“基础”资产
COIN_ID_MAP = {
    "ETH": "ethereum",
    "BTC": "bitcoin",
    "AAVE": "aave",
    "EURC": "euro-coin",
    "GHO": "gho",
    "USDC": "usd-coin",
    "USDbC": "bridged-usd-coin-base",
}

# 代理资产组
ETH_PROXY_SYMBOLS = ["WETH", "weETH", "ezETH", "cbETH", "wstETH", "wrsETH"]
BTC_PROXY_SYMBOLS = ["tBTC", "cbBTC", "LBTC"]

# 您在上一个脚本中定义的 CSV 输入目录
PRICE_DATA_DIR = Path(r"F:\Learning_journal_at_CUHK\FTEC5520_Appl Blockchain & Cryptocur\aave_data_collection\src\02_HF_simulation\02_data_collection_for_HF_simulation\01_collect_max_time_range_for_every_coin\data\every_icon_price_sequence")

# --- 新增：您指定的最终 CSV 输出目录 ---
FINAL_OUTPUT_DIR = Path(r"F:\Learning_journal_at_CUHK\FTEC5520_Appl Blockchain & Cryptocur\Simulation\R1\data")

# --- 2. 配置 CoinGecko API ---
API_KEY = 'CG-xtrHV9nWxQxtwq8tyThUtjNC' 
HEADERS = {
    'accept': 'application/json',
    'x-cg-demo-api-key': API_KEY
}
CG_BASE_URL = "https://api.coingecko.com/api/v3"


# -----------------------------------------------------------------------------
# 步骤 2: 您的 GraphQL 响应数据 (固定的数量快照)
# -----------------------------------------------------------------------------
# 这是您在上一个请求中提供的 JSON 响应
response_data = {
  "data": {
    "reserves": [
      {
        "decimals": 18, "price": {"id": "0x04c0599ae5a44757c0af6f9ec3b93da8976c150a", "priceInEth": "0"},
        "reserveLiquidationThreshold": "7700", "symbol": "weETH", "totalCurrentVariableDebt": "5711413226611483813",
        "totalLiquidity": "101060475476073517363523", "totalPrincipalStableDebt": "0"
      },
      {
        "decimals": 18, "price": {"id": "0x236aa50979d5f3de3bd1eeb40e81137f22ab794b", "priceInEth": "9691461317600"},
        "reserveLiquidationThreshold": "7800", "symbol": "tBTC", "totalCurrentVariableDebt": "2319486623176828625",
        "totalLiquidity": "10340642374716541723", "totalPrincipalStableDebt": "0"
      },
      {
        "decimals": 18, "price": {"id": "0x2416092f143378750bb29b79ed961ab195cceea5", "priceInEth": "0"},
        "reserveLiquidationThreshold": "10", "symbol": "ezETH", "totalCurrentVariableDebt": "0",
        "totalLiquidity": "62724757656228962796", "totalPrincipalStableDebt": "0"
      },
      {
        "decimals": 18, "price": {"id": "0x2ae3f1ec7f1f5012cfeab0185bfc7aa3cf0dec22", "priceInEth": "0"},
        "reserveLiquidationThreshold": "7900", "symbol": "cbETH", "totalCurrentVariableDebt": "73211718930522736606",
        "totalLiquidity": "7275725911966508937708", "totalPrincipalStableDebt": "0"
      },
      {
        "decimals": 18, "price": {"id": "0x4200000000000000000000000000000000000006", "priceInEth": "322840000000"},
        "reserveLiquidationThreshold": "8300", "symbol": "WETH", "totalCurrentVariableDebt": "111935033899531092134493",
        "totalLiquidity": "135773819228238313273787", "totalPrincipalStableDebt": "0"
      },
      {
        "decimals": 6, "price": {"id": "0x60a3e35cc302bfa44cb288bc5a4f316fdb1adb42", "priceInEth": "0"},
        "reserveLiquidationThreshold": "7800", "symbol": "EURC", "totalCurrentVariableDebt": "14738394184678",
        "totalLiquidity": "28137923993177", "totalPrincipalStableDebt": "0"
      },
      {
        "decimals": 18, "price": {"id": "0x63706e401c06ac8513145b7687a14804d17f814b", "priceInEth": "18186543000"},
        "reserveLiquidationThreshold": "6500", "symbol": "AAVE", "totalCurrentVariableDebt": "0",
        "totalLiquidity": "5743469788964741943928", "totalPrincipalStableDebt": "0"
      },
      {
        "decimals": 18, "price": {"id": "0x6bb7a212910682dcfdbd5bcbb3e28fb4e8da10ee", "priceInEth": "0"},
        "reserveLiquidationThreshold": "0", "symbol": "GHO", "totalCurrentVariableDebt": "7831765623389494312280982",
        "totalLiquidity": "14562490328570051552693819", "totalPrincipalStableDebt": "0"
      },
      {
        "decimals": 6, "price": {"id": "0x833589fcd6edb6e08f4c7c32d4f71b54bda02913", "priceInEth": "99986468"},
        "reserveLiquidationThreshold": "7800", "symbol": "USDC", "totalCurrentVariableDebt": "270641188114918",
        "totalLiquidity": "394263161444332", "totalPrincipalStableDebt": "0"
      },
      {
        "decimals": 18, "price": {"id": "0xc1cba3fcea344f92d9239c08c0568f6f2f0ee452", "priceInEth": "0"},
        "reserveLiquidationThreshold": "7900", "symbol": "wstETH", "totalCurrentVariableDebt": "6871703749728494496329",
        "totalLiquidity": "27527287398506498472580", "totalPrincipalStableDebt": "0"
      },
      {
        "decimals": 8, "price": {"id": "0xcbb7c0000ab88b473b1f5afd9ef808440eed33bf", "priceInEth": "8155695500000"},
        "reserveLiquidationThreshold": "7800", "symbol": "cbBTC", "totalCurrentVariableDebt": "40231161143",
        "totalLiquidity": "253922858089", "totalPrincipalStableDebt": "0"
      },
      {
        "decimals": 6, "price": {"id": "0xd9aaec86b65d86f6a7b5b1b0c42ffa531710b6ca", "priceInEth": "100010000"},
        "reserveLiquidationThreshold": "7800", "symbol": "USDbC", "totalCurrentVariableDebt": "403423125522",
        "totalLiquidity": "437671384568", "totalPrincipalStableDebt": "0"
      },
      {
        "decimals": 8, "price": {"id": "0xecac9c5f704e954931349da37f60e39f515c11c1", "priceInEth": "10849218252036"},
        "reserveLiquidationThreshold": "7300", "symbol": "LBTC", "totalCurrentVariableDebt": "0",
        "totalLiquidity": "5859006953", "totalPrincipalStableDebt": "0"
      },
      {
        "decimals": 18, "price": {"id": "0xedfa23602d0ec14714057867a78d01e94176bea0", "priceInEth": "0"},
        "reserveLiquidationThreshold": "10", "symbol": "wrsETH", "totalCurrentVariableDebt": "0",
        "totalLiquidity": "6478564545149594243854", "totalPrincipalStableDebt": "0"
      }
    ]
  }
}

# -----------------------------------------------------------------------------
# 步骤 3: CoinGecko 价格获取函数 (您已验证过)
# -----------------------------------------------------------------------------

def get_current_prices_usd(coin_id_map: dict) -> dict:
    """
    (无日志版)
    从 CoinGecko 获取 COIN_ID_MAP 中所有资产的当前 USD 价格。
    返回一个 {Symbol: Price_USD} 格式的字典。
    """
    ids_string = ",".join(coin_id_map.values())
    url = f"{CG_BASE_URL}/simple/price"
    params = { 'ids': ids_string, 'vs_currencies': 'usd' }
    
    try:
        response = requests.get(url, headers=HEADERS, params=params)
        response.raise_for_status() 
        cg_data = response.json()
        
        price_map_usd = {}
        for symbol, coin_id in coin_id_map.items():
            if coin_id in cg_data and 'usd' in cg_data[coin_id]:
                price = cg_data[coin_id]['usd']
                price_map_usd[symbol] = Decimal(str(price)) 
            else:
                # 即使找不到也只记录警告，不中断
                print(f"警告: CoinGecko 未返回 {symbol} 的价格。")
                
        return price_map_usd
        
    except Exception as e:
        print(f"严重错误: CoinGecko API 请求失败或处理时出错: {e}")
        return None

# -----------------------------------------------------------------------------
# 步骤 4: R1 计算函数 (已修改，更具稳健性)
# -----------------------------------------------------------------------------

def calculate_r1_proxy_strategy(reserves_data: list, prices_usd_map: dict):
    """
    (无日志版, 稳健版)
    根据 Aave Subgraph 数量 和 CoinGecko USD 价格计算 R_I 指标。
    使用 ETH 和 BTC 价格作为其衍生品的代理。
    如果 ETH 或 BTC 价格缺失，会跳过相应资产，而不是停止整个计算。
    """
    
    numerator_R1 = Decimal(0) 
    denominator_R1 = Decimal(0) 

    # 价格现在可能为 None，我们在循环内部检查
    ETH_PRICE_USD = prices_usd_map.get("ETH")
    BTC_PRICE_USD = prices_usd_map.get("BTC")

    for asset in reserves_data:
        symbol = asset.get('symbol')
        price_in_usd = None
        
        if symbol in STABLECOIN_SYMBOLS:
            price_in_usd = prices_usd_map.get(symbol)
        elif symbol in ETH_PROXY_SYMBOLS:
            price_in_usd = ETH_PRICE_USD # 可能为 None
        elif symbol in BTC_PROXY_SYMBOLS:
            price_in_usd = BTC_PRICE_USD # 可能为 None
        else:
            price_in_usd = prices_usd_map.get(symbol)
            
        # --- 这是核心修复 ---
        # 如果价格为 None (无论是直接找不到还是代理价格找不到)，
        # 我们就跳过这个资产，继续下一个。
        if not price_in_usd or asset.get('decimals') is None:
            continue
            
        decimals = int(asset['decimals'])
        divisor = Decimal(10) ** decimals
        
        try:
            D_j_amount = Decimal(asset['totalLiquidity']) / divisor
            B_j_amount = (Decimal(asset['totalPrincipalStableDebt']) +
                          Decimal(asset['totalCurrentVariableDebt'])) / divisor
            tau_j = Decimal(asset['reserveLiquidationThreshold']) / Decimal(10000)
        except Exception as e:
            continue
            
        if symbol in STABLECOIN_SYMBOLS:
            B_j_value_usd = B_j_amount * price_in_usd
            numerator_R1 += B_j_value_usd
        else:
            if tau_j == 0: 
                continue
            net_deposits_amount = D_j_amount - B_j_amount
            weighted_value_usd = (net_deposits_amount * tau_j) * price_in_usd
            denominator_R1 += weighted_value_usd

    if denominator_R1 <= 0:
        # 如果（在过滤后）分母仍为0，返回 'inf'
        return float('inf')
    
    R_I = numerator_R1 / denominator_R1
    return float(R_I)

# -----------------------------------------------------------------------------
# 步骤 5: 方案核心 - 加载历史价格、运行仿真并保存 CSV
# -----------------------------------------------------------------------------

if __name__ == "__main__":
    
    print("--- 步骤 1: 准备 R1 历史仿真 ---")
    
    # 确保最终输出目录存在
    FINAL_OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
    print(f"最终输出目录已准备好: {FINAL_OUTPUT_DIR}")

    # 1. 数据准备 (价格): 加载并合并所有基础资产的 CSV
    print(f"正在从 {PRICE_DATA_DIR} 加载价格 CSVs...")
    base_asset_dfs = {}
    base_assets_to_load = list(COIN_ID_MAP.keys())
    
    # --- 新增：检查 BTC 文件是否确实缺失 ---
    btc_file_path = PRICE_DATA_DIR / "BTC_hourly_price_data.csv"
    if not btc_file_path.exists():
        print("---")
        print("警告: 未找到 'BTC_hourly_price_data.csv'。")
        print("仿真将继续，但所有 BTC 衍生品 (tBTC, cbBTC, LBTC) 将被跳过。")
        print("要修复此问题，请将 '\"BTC\": \"bitcoin\"' 添加到您的 *下载脚本* 的 COIN_ID_MAP 中并重新运行。")
        print("---")
    # -------------------------------------

    for symbol in base_assets_to_load:
        file_path = PRICE_DATA_DIR / f"{symbol}_hourly_price_data.csv"
        if file_path.exists():
            df = pd.read_csv(file_path)
            # 修复：明确告知 Pandas 格式为 ISO8601
            df['datetime_utc'] = pd.to_datetime(df['datetime_utc'], format='ISO8601')
            df = df.set_index('datetime_utc')
            base_asset_dfs[symbol] = df.rename(columns={'price_usd': symbol})
        else:
            print(f"  -> 提示: 找不到价格文件 {file_path} (资产: {symbol})。")

    if not base_asset_dfs:
        print("!!! 严重错误: 未加载到任何价格数据，无法进行仿真。!!!")
    else:
        # 合并所有 DataFrame
        merged_prices_df = pd.concat(base_asset_dfs.values(), axis=1, join='outer')
        merged_prices_df = merged_prices_df.sort_index()
        
        # 填充缺失值
        merged_prices_df = merged_prices_df.ffill().bfill()
        
        # 过滤到您指定的时间范围
        start_time_str = "2025-08-01T14:04:37.234000+00:00"
        end_time_str = "2025-10-30T13:01:57.236000+00:00"
        
        merged_prices_df = merged_prices_df.loc[start_time_str:end_time_str]
        
        print(f"价格数据已合并并过滤。共 {len(merged_prices_df)} 个小时数据点。")

        # 2. 数据准备 (数量): 
        reserves_list = response_data.get("data", {}).get("reserves", [])
        if not reserves_list:
            print("!!! 严重错误: 'response_data' 中没有 'reserves' 列表。")
        else:
            # 3. 仿真计算
            print("正在按小时遍历价格，仿真 R1 波动...")
            simulation_results = []
            
            for timestamp, price_row in tqdm(merged_prices_df.iterrows(), total=len(merged_prices_df)):
                
                # --- 这是核心修复 ---
                # 我们不再因为缺少 BTC 而跳过整个时间点。
                # 我们只在缺少 ETH (主要抵押品代理) 时才跳过。
                
                prices_usd_map_hourly = {}
                for symbol in COIN_ID_MAP.keys():
                    if symbol in price_row and pd.notna(price_row[symbol]):
                        prices_usd_map_hourly[symbol] = Decimal(str(price_row[symbol]))
                
                # 稳健性检查：我们必须有 ETH 价格才能计算分母的大部分
                if "ETH" not in prices_usd_map_hourly:
                    print(f"警告：在 {timestamp} 缺少 ETH 的价格，将跳过此时间点")
                    continue # 跳到下一个时间戳
                
                # 调用 R1 计算 (新版稳健函数)
                r1_hourly = calculate_r1_proxy_strategy(reserves_list, prices_usd_map_hourly)
                
                if r1_hourly is not None:
                    simulation_results.append({
                        "datetime_utc": timestamp,
                        "R1_simulated": r1_hourly
                    })
            
            print("仿真计算完成。")

            # 4. 结果分析与保存
            if not simulation_results:
                print("!!! 错误: 仿真未产生任何结果。请检查价格数据和时间范围。")
            else:
                r1_df = pd.DataFrame(simulation_results)
                # (修改) 在保存前，我们重置索引，使 'datetime_utc' 成为一个列
                r1_df = r1_df.reset_index()

                print("\n--- R1 仿真结果统计 ---")
                print(r1_df['R1_simulated'].describe())
                
                # --- 修改：保存到 CSV 而不是绘图 ---
                final_output_file = FINAL_OUTPUT_DIR / "r1_simulation_results.csv"
                try:
                    # (修改) 设置 index=False，因为日期已是普通列
                    r1_df.to_csv(final_output_file, index=False, encoding='utf-8')
                    print(f"\n--- 仿真结果已成功保存 ---")
                    print(f"文件路径: {final_output_file}")
                except Exception as e:
                    print(f"\n!!! 错误: 无法保存 CSV 文件到 {final_output_file}")
                    print(f"错误信息: {e}")
```
